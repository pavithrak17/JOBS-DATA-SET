```python
E={
    'Name':[],
    'Salary':[],
    'Designation':[],
    'Years of Experience':[],
    'no of lang_known':[]
}
E
```

```python
print('Employee details')
while(True):
  name=input('Name:')
  E['Name'].append(name)
  Sal=int(input('Salary:'))
  E['Salary'].append(Sal)
  Des=input('Designation:')
  E['Designation'].append(Des)
  Yoe=int(input('years of exp:'))
  E['Years of Experience'].append(Yoe)
  lang=int(input('number of lang known:'))
  E['no of lang_known'].append(lang)
  n=input('Do you want to continue? Yes or No :')
  if (n=='Yes'):
    print('please continue entering the details:')

  if(n=='No'):
   break
```
